```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from underlyings.stock import Stock
from underlyings.index import Index
from datetime import datetime

%matplotlib inline
spx = Index("SPX")
apple = Stock('AAPL')
tsla = Stock("TSLA")

start_date =  pd.to_datetime('2018-01-01')
end_date = datetime.today().strftime('%Y-%m-%d')

df_apple = apple.get_quotes(start_date, end_date)
df_tsla = tsla.get_quotes(start_date, end_date)
df_spx = spx.get_quotes(start_date, end_date)

plt.plot(df_apple, label='AAPL')
plt.plot(df_tsla, label='TSLA')

# Setting the title and labels
plt.title('TSLA & AAPL Series')
plt.xlabel('X')
plt.ylabel('Price')

# Adding the legend
plt.legend()
```
