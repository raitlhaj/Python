```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from underlyings.stock import Stock
from underlyings.index import Index
from datetime import datetime

%matplotlib inline
spx = Index("SPX")
stock_symbols = ['AAPL', 'TSLA', 'NVDA', 'AVGO', 'IBM', 'MRVL', 'MSFT', 'AMZN']

start_date =  pd.to_datetime('2023-01-02')
end_date = datetime.today().strftime('%Y-%m-%d')

for symbol in stock_symbols:
    
    df_stock = Stock(symbol).get_quotes(start_date, end_date) 
    df_stock = df_stock.div(df_stock.iloc[0])
    df_stock = df_stock.apply( lambda x: x -1)
    
    plt.plot(df_stock, label=symbol)

# Setting the title and labels
plt.title('Taux Forwards: Rf')
plt.xlabel('X')
plt.ylabel('Price')

# Adding the legend
plt.legend()
```
